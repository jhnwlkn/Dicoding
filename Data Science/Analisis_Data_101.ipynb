```python
# Dataset: https://raw.githubusercontent.com/jhnwlkn/Dicoding/main/Data%20Science/Bali%20Popular%20Destination%20for%20Tourist%202022%20-%20Sheet1.csv
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Wrangling

```python
df = pd.read_csv('https://raw.githubusercontent.com/jhnwlkn/Dicoding/main/Data%20Science/Bali%20Popular%20Destination%20for%20Tourist%202022%20-%20Sheet1.csv')
df.head(5)
```

```python
df_sorted = df.sort_values(by='Google Maps Rating', ascending=False)
df_sorted.head(5)
```

```python
avg_rating = df['Google Maps Rating'].mean()
print(avg_rating)
```

# Exploratory Data Analysis (EDA)

```python
df_sorted.drop_duplicates(inplace=True)
df_sorted.head(5)
```

```python
df_sorted = df_sorted.dropna()
df_sorted
```

```python
deleted_colom = ['Source', 'Coordinate']
df_clean = df.drop(deleted_colom, axis=1)
df_clean
```

```python
df_clean = df_clean[df_clean['Google Maps Rating'] < 4.5]
df_clean
```

```python
df_clean = df_clean.sort_values(by='Google Maps Rating', ascending=True)
correlation = df_clean['Google Maps Rating'].corr(df['Google Reviews (Count)'])
print(f'Nilai Korelasi: {correlation}')
```

```python
plt.scatter(df_clean['Google Maps Rating'], df_clean['Google Reviews (Count)'])

plt.xlabel('Google Maps Rating')
plt.ylabel('Google Reviews (Count)')

plt.title('Scatter Plot antara Google Maps Rating dan Google Review')
```
